```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd
import numpy as np

# numerical preprocessing
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
])

# categorical preprocessing
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# combine preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_cols),
        ('cat', categorical_transformer, cat_cols)
    ]
)

# models to compare
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42)
}

results = []

for name, model in models.items():
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    results.append({
        'Model': name,
        'MAE': mae,
        'RMSE': rmse,
        'R2 Score': r2
    })

results_df = pd.DataFrame(results).sort_values(by='R2 Score', ascending=False)
results_df
```

```python
!pip install gradio -q

from google.colab import drive
drive.mount('/content/drive')

import os
import joblib
import pandas as pd
import numpy as np
import gradio as gr

# -------------------------------
# Load model and artifact info
# -------------------------------
save_dir = "/content/drive/MyDrive/melbourne_saved_artifacts"

model = joblib.load(f"{save_dir}/best_rf_model.pkl")
artifact_info = joblib.load(f"{save_dir}/artifact_info.pkl")

print("Loaded files from:", save_dir)
print("Available files:", os.listdir(save_dir))


# -------------------------------
# Feature engineering function
# -------------------------------
def engineer_features(df):
    df = df.copy()

    # Parse Date
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce", dayfirst=True)

    # Extract year and month
    df["SaleYear"] = df["Date"].dt.year
    df["SaleMonth"] = df["Date"].dt.month

    # Safe fill if dates are missing
    if "saleyear_min_used_for_propertyage" in artifact_info:
        base_year = artifact_info["saleyear_min_used_for_propertyage"]
    else:
        base_year = df["SaleYear"].dropna().min() if df["SaleYear"].notna().any() else 2016

    df["SaleYear"] = df["SaleYear"].fillna(base_year)
    df["SaleMonth"] = df["SaleMonth"].fillna(1)

    # Numeric cleanup
    for col in ["Rooms", "Bathroom", "Car", "Landsize", "Distance", "Bedroom2", "Propertycount"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    # Avoid divide by zero
    rooms_safe = df["Rooms"].replace(0, np.nan) if "Rooms" in df.columns else np.nan

    # Engineered features
    if {"Landsize", "Rooms"}.issubset(df.columns):
        df["PricePerRoomProxy"] = df["Landsize"] / rooms_safe
    if {"Bathroom", "Rooms"}.issubset(df.columns):
        df["BathPerRoom"] = df["Bathroom"] / rooms_safe
    if {"Car", "Rooms"}.issubset(df.columns):
        df["CarPerRoom"] = df["Car"] / rooms_safe
    if "SaleYear" in df.columns:
        df["PropertyAge"] = df["SaleYear"] - base_year
    if "Landsize" in df.columns:
        df["LogLandsize"] = np.log1p(df["Landsize"].clip(lower=0))

    # Fill engineered NaNs
    for col in ["PricePerRoomProxy", "BathPerRoom", "CarPerRoom", "PropertyAge", "LogLandsize"]:
        if col in df.columns:
            df[col] = df[col].replace([np.inf, -np.inf], np.nan)
            df[col] = df[col].fillna(df[col].median() if df[col].notna().any() else 0)

    return df


# -------------------------------
# Prediction function
# -------------------------------
def predict_from_csv(file):
    try:
        df = pd.read_csv(file.name)
        original_df = df.copy()

        required_cols = [
            "Suburb", "Rooms", "Type", "Method", "SellerG",
            "Distance", "Postcode", "Bedroom2", "Bathroom", "Car",
            "Landsize", "CouncilArea", "Lattitude", "Longtitude",
            "Regionname", "Propertycount", "Date"
        ]

        missing_cols = [col for col in required_cols if col not in df.columns]
        if missing_cols:
            return None, f"Missing required columns: {missing_cols}"

        # Apply feature engineering
        df = engineer_features(df)

        # Match training columns if saved
        if "model_input_columns" in artifact_info:
            needed_cols = artifact_info["model_input_columns"]

            for col in needed_cols:
                if col not in df.columns:
                    df[col] = np.nan

            df = df[needed_cols]

        # Predict
        preds_log = model.predict(df)
        preds_price = np.expm1(preds_log)

        # Add output column
        original_df["PredictedPrice"] = np.round(preds_price, 2)

        # Save output
        output_path = "/content/predicted_house_prices.csv"
        original_df.to_csv(output_path, index=False)

        return output_path, "Prediction completed successfully."

    except Exception as e:
        return None, f"Error: {str(e)}"


# -------------------------------
# Gradio GUI
# -------------------------------
demo = gr.Interface(
    fn=predict_from_csv,
    inputs=gr.File(label="Upload Input CSV"),
    outputs=[
        gr.File(label="Download Predicted CSV"),
        gr.Textbox(label="Status")
    ],
    title="Melbourne House Price Prediction GUI",
    description="Upload a CSV file with house features and get predicted prices."
)

demo.launch(share=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'path' is defined (it comes from cell xg40v4IfguJX)
# Reload the 'house' DataFrame and apply initial cleaning steps
# to ensure it's in the correct state for this visualization.
if 'path' not in globals():
    # This part should ideally be handled by running prior cells
    # But for self-containment if this cell is run independently:
    print("Warning: 'path' not found. Attempting to re-download dataset...")
    import kagglehub
    path = kagglehub.dataset_download("anthonypino/melbourne-housing-market")

house = pd.read_csv(f"{path}/Melbourne_housing_FULL.csv")

# Apply preprocessing steps relevant before this missing value visualization
house = house.dropna(subset=['Price']) # From cell yUELvmfskAxK
house = house.drop(['BuildingArea', 'YearBuilt'], axis=1) # From cell pMAjN0z3kHo3

# Calculate missing values
missing_data_count = house.isnull().sum().sort_values(ascending=False)

# Filter out columns with 0 missing values for better visualization
missing_data_count = missing_data_count[missing_data_count > 0]

if not missing_data_count.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=missing_data_count.index, y=missing_data_count.values, palette='viridis', hue=missing_data_count.index, legend=False) # Fixed FutureWarning
    plt.title('Count of Missing Values per Column (After Initial Cleaning)')
    plt.xlabel('Columns')
    plt.ylabel('Number of Missing Values')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print('No missing values found in the DataFrame after initial cleaning.')
```
